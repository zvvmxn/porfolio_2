# 1. 서론

## 1.1. 연구 배경 및 목적

현대 사회에서 공공자전거는 친환경 교통수단으로 자리 잡으며, 다양한 계층에게 중요한 이동 및 여가 수단으로 인식되고 있습니다. 그러나 노년층의 경우, 운동과 여가를 위한 공공자전거 이용 의사가 있음에도 불구하고 실제 이용률은 여전히 낮은 상황입니다.

본 분석은 노년층이 단순 이동을 넘어 건강 증진과 신체 활동의 일환으로 공공자전거를 적극적으로 활용하고자 하는 수요가 존재함에도 불구하고, 그 수요가 실제 이용률로 이어지지 않는 현상의 원인을 규명하고자 하는 문제의식에서 출발했습니다.

이를 위해, 노년층의 낮은 이용건수가 이용 욕구 부족 때문인지, 혹은 이용 과정에서의 장벽(인지, 접근성 등)에 기인하는지를 다각적으로 분석하고, 노년층의 공공자전거 이용률을 높이기 위한 실질적 개선 방향을 모색하고자 합니다.

## 1.2. 연구 질문 및 가설 설정

### 1.2.1. 연구 질문
본 연구는 다음과 같은 핵심 질문을 중심으로 진행됩니다.
> 1. 노년층의 따릉이 이용건수가 낮은 원인은 무엇인가?
> 2. 노년층은 따릉이에 대해 여가적·운동적 목적의 이용 욕구를 가지고 있는가?
> 3. 노년층은 따릉이를 이용하는 과정에서 접근성의 문제를 겪고 있는가?

### 1.2.2. 연구 가설
이러한 연구 질문에 따라 다음과 같은 세 가지 주요 가설을 설정하였습니다.

> 1. 인지 가설: 노년층은 따릉이에 대해 잘 알지 못해 이용하지 않는다.
> 2. 욕구 가설: 노년층은 따릉이를 여가나 운동 목적으로 이용할 필요성을 느끼지 않는다.
> 3. 접근성 가설: 노년층은 따릉이를 이용하는 과정에서 물리적·심리적 장벽을 경험한다.

본 연구는 욕구 가설과 접근성 가설을 정량적 데이터 분석을 통해 검증하고, 두 가설이 기각될 경우 인지 가설을 채택하는 방식으로 진행됩니다.

## 3.5. `연령대` 열

### 3.5.1. 연령대 열 추가

아래 기준에 기반해 연령대를 구분합니다.
-  노년: 1959년생(만 65세) ~ 1939년생(만 85세)
-  중년: 1984년생(만 40세) ~ 1960년생(만 60세)
-  청년: 2005년생(만 19세) ~ 1985년생(만 39세)

```python
# 연령대 열 생성
def categorize_age(year):
    if 1939 <= year <= 1959:
        return '노년'
    elif 1960 <= year <= 1984:
        return '중년'
    elif 1985 <= year <= 2005:
        return '청년'
    else:
        return '기타'

# 연령대 열 추가
all_df['연령대'] = all_df['생년'].apply(categorize_age)
all_df.head()
```

```python
# '생년' 열 삭제
all_df.drop(columns=['생년'], inplace=True)
all_df.head()
```

첫째, 노년층의 구분은 국민연금 지급 기준인 만 65세 이상을 기준으로 설정했습니다. 이는 국내에서 노년층을 정의하는 대표적인 사회적·정책적 기준으로, 다양한 복지 정책과 통계 분석에서도 동일한 기준을 적용하고 있습니다.

둘째, 노년층을 중심으로 분석하는 본 연구의 목적상, 나머지 성인 집단은 청년층과 중년층의 두 범주로 통합했습니다. 일반적으로 성인의 생애 주기는 청년기, 중년기, 장년기, 노년기로 4단계로 구분되지만, 본 분석은 노년층의 특성을 비교·분석하는 데 초점을 맞추었기 때문에 비(非)노년층을 세분화하기보다는 분석의 일관성과 해석의 용이성을 높이기 위해 3단계 연령 구분 체계를 채택했습니다.

셋째, 이러한 구분 방식은 노년층과 비(非)노년층 간의 차이를 보다 명확하게 도출하고, 중년과 장년을 통합하여 생애주기 연속성 내에서 연령대 간 경향성을 포착하는 데 유리하다는 점에서도 타당성을 갖습니다.

따라서, 본 분석에서의 연령 구분은 노년층의 공공자전거 이용 행태를 다른 성인 집단과 비교하는 과정에서 분석적 명료성을 유지하고, 정책적 시사점을 도출하는 데 최적화된 기준이라 할 수 있습니다.

## 3.7. `자치구`, `위도`, `경도` 열 추가

```python
# 자치구, 위도, 경도 데이터 추가

# '대여 대여소번호'와 '반납 대여소번호'를 문자열(str)로 변환
all_df['대여 대여소번호'] = all_df['대여 대여소번호'].astype(str)
all_df['반납 대여소번호'] = all_df['반납 대여소번호'].astype(str)
stations['대여소번호'] = stations['대여소번호'].astype(str)

# 대여 대여소 정보 병합
all_df = all_df.merge(stations[['대여소번호', '자치구', '위도', '경도']],
              left_on='대여 대여소번호',
              right_on='대여소번호',
              how='left')

# 열 이름 변경 (대여소 기준)
all_df.rename(columns={
    '자치구': '대여 대여소 자치구',
    '위도': '대여 대여소 위도',
    '경도': '대여 대여소 경도'
}, inplace=True)
all_df.drop(columns=['대여소번호'], inplace=True)

# 반납 대여소 정보 병합
all_df = all_df.merge(stations[['대여소번호', '자치구', '위도', '경도']],
              left_on='반납 대여소번호',
              right_on='대여소번호',
              how='left')

# 열 이름 변경 (반납소 기준)
all_df.rename(columns={
    '자치구': '반납 대여소 자치구',
    '위도': '반납 대여소 위도',
    '경도': '반납 대여소 경도'
}, inplace=True)
all_df.drop(columns=['대여소번호'], inplace=True)

all_df
```

### 4.2.4. 이상치 제거

```python
# 상위 1% 기준 찾기
time_threshold = eda_df['이용시간(분)'].quantile(0.99)
distance_threshold = eda_df['이용거리(M)'].quantile(0.99)

print(f"이용시간 상위 1% 기준: {time_threshold:.2f}분")
print(f"이용거리 상위 1% 기준: {distance_threshold:.2f}m")

# 보수적 이상치 제거
filtered_df = eda_df[
    (eda_df['이용시간(분)'] <= time_threshold) &
    (eda_df['이용거리(M)'] <= distance_threshold)
]

# 데이터 손실 확인
total_count = len(eda_df)
filtered_count = len(filtered_df)
lost_count = total_count - filtered_count
lost_percentage = (lost_count / total_count) * 100

print(f"전체 데이터 수: {total_count:,}")
print(f"이상치 제거 후 데이터 수: {filtered_count:,}")
print(f"제거된 데이터 수: {lost_count:,} ({lost_percentage:.2f}%)")
```

# 5. 데이터 분석 - 욕구 가설 검증

## 5.1. 추가 이용량

### 5.1.1. 이용건수만으로는 욕구를 판단할 수 없는 이유
노년층의 공공자전거 이용건수가 다른 연령대에 비해 낮다는 점은 이미 확인되었습니다. 그러나 이용건수만으로는 이용 의사(욕구)를 직접 평가하기 어렵습니다. 이용건수는 다음과 같이 인지, 욕구, 접근성이라는 복합적인 요인에 의해 결정되기 때문입니다.

> 1. 인지: 따릉이의 존재와 이용 방법을 알지 못하면 이용이 불가능합니다.
> 2. 욕구: 따릉이를 알고 있어도 이용할 이유가 없다면 행동으로 이어지지 않습니다.
> 3. 접근성: 이용 욕구가 있어도 대여소 접근성이나 사용법의 장벽이 크면 이용이 제한됩니다.

따라서, 이용건수만으로는 욕구가 부족해서 이용하지 않는 것인지, 접근성 문제로 이용하지 못하는 것인지 명확히 구분할 수 없습니다.

### 5.1.2. `이용시간(분)`과 `이용거리(M)`의 한계
그렇다면 이용시간과 이용거리를 직접 비교하면 욕구를 알 수 있을까요? 문제는 대여소 간 물리적 거리가 이용시간과 이용거리에 큰 영향을 미친다는 것입니다. 예를 들어, 이동 목적의 대여라도 대여소 간 거리가 길면 자연히 이용시간과 이용거리가 늘어날 수밖에 없습니다.

따라서, 단순 이용시간과 이용거리만으로는 "이용 목적"을 판단하는 데 한계가 있습니다.

### 5.1.3. "추가 이용량" 도입의 필요성
이러한 한계를 극복하고, 노년층의 따릉이 이용이 단순 이동을 넘어 여가나 운동 목적을 포함하는지를 분석하기 위해, "추가 이용량"을 도입합니다.

- `추가 이용시간(분)`: 두 대여소 간 예상 이동시간보다 얼마나 더 오래 이용했는지
- `추가 이용거리(M)`: 두 대여소 간 거리보다 얼마나 더 길게 이동했는지

추가 이용량은 대여소 간 거리라는 환경적 요인을 통제하고, 실제 이용자가 이동 목적 이상으로 따릉이를 얼마나 활용했는지를 측정하는 지표입니다. 따라서, 노년층의 추가 이용량이 다른 연령대보다 유의미하게 낮다면 욕구 부족을 시사하고, 반대로 높거나 유사하다면 여가와 운동을 포함한 이용 욕구가 존재한다는 강력한 근거가 됩니다.

### 5.1.4. "추가 이용량" 정의와 계산 방식
1. `추가 이용시간(분)`
    - 정의: 대여소 간 이동에 걸리는 예상 소요시간을 초과하여 자전거를 이용한 시간입니다.
    - 계산 공식:
        - 추가 이용시간(분) = 이용시간(분) − 예상 이동시간(분)
    - 세부 설명:
        - 예상 이동시간(분): 대여소 간 이동시간의 중앙값을 기준으로 산정
    - 해석:
        - 추가 이용시간이 0이면, 사용자가 최소한의 이동만 수행한 것으로 해석됩니다.
        - 추가 이용시간이 양수일 경우, 예상보다 더 오래 자전거를 이용했음을 의미하며, 여가적·운동적 목적을 시사할 수 있습니다.
2. `추가 이용거리(M)`
    - 정의: 대여소 간 예상 이동거리를 초과하여 실제로 이동한 거리입니다.
    - 계산 방식:
        - 추가 이용거리(M) = 이용거리(M) − 보정 이동거리(M)
    - 세부 설명:
        - 보정 이동거리(M): 대여소 간 직선 거리를 현실적인 이동 경로를 고려하여 1.4배로 보정한 값
        - 보정값(1.4배):
            - 대도시권(Metropolitan area)의 우회계수는 1.238에서 1.465까지 범위에 분포하고 평균은 1.339로 분석되었습니다.
            - 우회계수(Circuity) = “현실 최단 이동 거리 / 직선 거리(유클리드 거리)”로 계산되는, 직선 거리 대비 실제 이동 거리의 비율을 나타냅니다.
                - 출처: Giacomin, David & Levinson, David. (2015). Road network circuity in metropolitan areas. Environment and Planning B: Planning and Design.
    - 해석:
        - 추가 이용거리가 0 이하라면, 단순 이동 목적에 그쳤음을 의미합니다.
        - 추가 이용거리가 양수일 경우, 경로를 우회하거나 추가 이동을 수행했다는 뜻으로, 여가·운동 목적의 가능성을 시사합니다.

### 5.1.5. 추가 이용량 계산

```python
# 데이터 복사 및 확인
need_df = filtered_df.copy()
need_df.head()
```

```python
# 불필요한 열 목록
columns_to_drop = ['날짜', '시간', '대여 대여소 자치구', '반납 대여소 자치구']

# 불필요한 열 삭제
need_df.drop(columns=columns_to_drop, inplace=True)

# 결과 확인
need_df.head()
```

```python
# 1. '이용시간(분)'의 중앙값 계산
median_travel_time = need_df['이용시간(분)'].median()

# 2. 추가 이용시간(분) 열 생성 (음수는 0으로 처리)
need_df['추가 이용시간(분)'] = (need_df['이용시간(분)'] - median_travel_time).apply(lambda x: max(0, x))

# 3. 결과 확인
print(f"이용시간 중앙값: {median_travel_time}분")
need_df[['연령대', '이용시간(분)', '추가 이용시간(분)']].head()
```

```python
# 1. 대여소 간 직선거리(M) 계산 함수
def calculate_distance(row):
    start = (row['대여 대여소 위도'], row['대여 대여소 경도'])
    end = (row['반납 대여소 위도'], row['반납 대여소 경도'])
    return geodesic(start, end).meters

# 2. 대여소 간 거리 계산 및 보정
need_df['예상 이동거리(M)'] = need_df.apply(calculate_distance, axis=1) # 직선 이동거리
correction_factor = 1.4  # 현실 반영 보정값 (1.3~1.5 권장)
need_df['보정 이동거리(M)'] = (need_df['예상 이동거리(M)'] * correction_factor).round(2)

# 3. 추가 이용거리(M) 계산 (음수는 0으로 처리)
need_df['추가 이용거리(M)'] = (need_df['이용거리(M)'] - need_df['보정 이동거리(M)']).apply(lambda x: max(0, x))

# 4. 결과 확인
need_df[['연령대', '이용거리(M)', '예상 이동거리(M)', '보정 이동거리(M)', '추가 이용거리(M)']].head()
```

```python
# 불필요한 열 목록
columns_to_drop = ['대여 대여소번호', '대여 대여소 위도', '대여 대여소 경도', '반납 대여소번호', '반납 대여소 위도',
       '반납 대여소 경도', '이용시간(분)', '이용거리(M)', '예상 이동거리(M)', '보정 이동거리(M)']

# 불필요한 열 삭제
need_df.drop(columns=columns_to_drop, inplace=True)
need_df.head()
```

        - 플롯 상에서 데이터 점들이 45도 직선(정규성 기준선)을 따라 일관되게 분포하지 않고, 특히 양 끝에서 크게 벗어나는 모습을 보입니다.
        - 이는 긴 꼬리 분포(long-tail distribution)의 형태를 시사하며, 정규성을 충족하지 않을 가능성이 높음을 의미합니다.
    2. `추가 이용거리(M)`
        - 추가 이용거리 역시 직선과 큰 괴리를 보이며, 특히 극단값(outliers)이 분포의 왜곡을 초래하는 모습입니다.
        - 양측 끝에서 크게 벗어난 점들은 비정규성(non-normality)을 강하게 시사합니다.
- 잠정적 결론 및 다음 단계
    - 정규성 가정이 시각적으로 충족되지 않을 가능성이 높습니다.
    - 그러나 Q-Q 플롯은 시각적 참고 자료일 뿐이며, 통계적 검정 없이 최종 결론을 내릴 수 없습니다.
    - 다음은 Anderson-Darling Test를 통해 정규성 여부를 통계적으로 검정합니다.

### 5.3.2. Anderson-Darling Test

```python
# Anderson-Darling Test 실행 함수
def anderson_test(data, column):
    result = anderson(data[column], dist='norm')
    print(f"\n🚩 {column}의 Anderson-Darling Test 결과:")
    print(f"Test Statistic: {result.statistic:.4f}")
    print("임계값(Critical Values):", result.critical_values)
    print("유의 수준(Significance Levels):", result.significance_level)
    
    # 결과 해석
    for i in range(len(result.critical_values)):
        sl, cv = result.significance_level[i], result.critical_values[i]
        if result.statistic > cv:
            print(f"- {sl:.1f}% 수준에서 정규성을 기각합니다. (통계값: {result.statistic:.4f} > 임계값: {cv:.4f})")
        else:
            print(f"- {sl:.1f}% 수준에서 정규성을 기각하지 않습니다. (통계값: {result.statistic:.4f} ≤ 임계값: {cv:.4f})")

# 추가 이용시간(분)과 추가 이용거리(M) 검정
anderson_test(zero_filtered_df, '추가 이용시간(분)')
anderson_test(zero_filtered_df, '추가 이용거리(M)')
```

- 해석
    1. `추가 이용시간(분)`
        - 검정 통계량(Test Statistic): 218,337.2179
        - 임계값(Critical Values): [0.576, 0.656, 0.787, 0.918, 1.092]
        - 유의 수준(Significance Levels): [15%, 10%, 5%, 2.5%, 1%]
        - 해석:
            - 모든 유의 수준에서 Test Statistic(218,337.2179)이 임계값을 크게 초과했습니다.
            - 이는 1% 수준에서도 귀무가설(정규성을 따른다)을 강하게 기각함을 의미합니다.
            - 따라서 `추가 이용시간(분)`은 정규성을 따르지 않는다고 결론내릴 수 있습니다.
    2. `추가 이용거리(M)`
        - 검정 통계량(Test Statistic): 699,584.9351
        - 임계값(Critical Values): [0.576, 0.656, 0.787, 0.918, 1.092]
        - 유의 수준(Significance Levels): [15%, 10%, 5%, 2.5%, 1%]
        - 해석:
            - Test Statistic(699,584.9351)이 역시 모든 유의 수준의 임계값을 크게 초과했습니다.
            - 이는 1% 수준에서도 정규성을 강하게 기각하는 결과입니다.
            - `추가 이용거리(M)` 역시 정규성을 따르지 않는다는 결론을 내릴 수 있습니다.

### 5.3.3. 결론
1. Q-Q 플롯과 Anderson-Darling Test 결과 모두, 추가 이용시간과 추가 이용거리는 정규성을 따르지 않는다고 판명되었습니다.
2. 따라서, 이후 통계 검정에서는 비정규 분포를 가정한 방법을 선택해야 하며, 이는 Kruskal-Wallis Test와 Dunn’s 사후 검정을 진행하는 근거가 됩니다.

    - 이는 연령대 간의 추가 이용시간과 추가 이용거리의 차이가 통계적으로 유의미함을 의미합니다.
    - 즉, 청년, 중년, 노년층 간에 추가 이용량에 차이가 존재한다는 가설을 지지하는 결과입니다.
- 다음 단계:
    - Kruskal-Wallis Test는 차이의 존재 여부만 확인할 뿐, 어떤 그룹 간 차이가 유의미한지는 알려주지 않습니다.
    - 따라서, Dunn's 을 진행하여 어떤 연령대 간에 유의한 차이가 있는지를 확인해야 합니다.

### 5.6.2. Dunn's Test

```python
# 🕰️ 추가 이용시간(분) Dunn's Test
print("🚲 추가 이용시간(분) Dunn's Test 결과:")
dunn_time = sp.posthoc_dunn(
    zero_filtered_df, val_col='추가 이용시간(분)', group_col='연령대', p_adjust='bonferroni'
)
print(dunn_time)

# 📏 추가 이용거리(M) Dunn's Test
print("\n🚲 추가 이용거리(M) Dunn's Test 결과:")
dunn_distance = sp.posthoc_dunn(
    zero_filtered_df, val_col='추가 이용거리(M)', group_col='연령대', p_adjust='bonferroni'
)
print(dunn_distance)
```

- 해석:
    - `추가 이용시간(분)`:
        - 노년 vs 중년, 노년 vs 청년, 중년 vs 청년 모두 p-value = 0.0으로 유의미한 차이가 있습니다.
        - 이는 모든 연령대 간의 추가 이용시간에서 통계적으로 유의한 차이가 존재함을 의미합니다.
    - `추가 이용거리(M)`:
        - 노년 vs 중년, 노년 vs 청년, 중년 vs 청년 모두 p-value < 0.05로 유의미한 차이가 나타났습니다.
        - 특히 노년과 청년 간의 p-value는 2.586e-257, 중년과 청년 간의 p-value는 0.0로 매우 강한 통계적 유의성을 보입니다.
- 결론:
    - Dunn's Test 결과, 추가 이용시간과 추가 이용거리 모두에서 모든 연령대 간에 통계적으로 유의미한 차이가 확인되었습니다.
    - 이는 노년층의 추가 이용량이 다른 연령대와 명확하게 구분되는 특징을 보인다는 점을 시사합니다.
    - Games-Howell Test와의 일관된 결과로, 분석의 신뢰성을 높여줍니다.

## 5.7. 보완적 분석

### 5.7.1. Box-Cox 변환

```python
# Box-Cox 변환 대상 컬럼
time_col = '추가 이용시간(분)'
distance_col = '추가 이용거리(M)'

# 원본에 영향이 없도록 DataFrame 복사 (경고 방지)
box_cox_df = zero_filtered_df.copy()

# Box-Cox 변환 수행
time_transformed, time_lambda = stats.boxcox(box_cox_df[time_col])
distance_transformed, distance_lambda = stats.boxcox(box_cox_df[distance_col])

# 변환 결과 저장 (loc 사용)
box_cox_df.loc[:, '추가 이용시간(분)_BoxCox'] = time_transformed
box_cox_df.loc[:, '추가 이용거리(M)_BoxCox'] = distance_transformed

# 변환된 데이터 확인
box_cox_df.head(3)
```

- Box-Cox 변환은 정규성을 개선하고 분산을 안정화하기 위해 사용되는 비선형 변환 방법입니다.
    - Box-Cox 변환 목적:
        - 정규성 확보: 데이터의 왜도를 줄이고 정규 분포에 가깝게 만들어, 정규성 기반 검정을 수행할 수 있도록 합니다.
        - 분산 안정화: 데이터의 분포 폭을 일정하게 맞춰, 분석 결과의 신뢰성을 높입니다.
        - 추가 분석 수행: 정규성 기반 분석 방법(ANOVA, Tukey's HSD 등)을 적용할 수 있게 됩니다.
- 본 분석에서는 추가 이용시간(분)과 추가 이용거리(M) 변수를 대상으로 Box-Cox 변환을 수행했습니다.
- 변환 후, 새롭게 생성된 `추가 이용시간(분)_BoxCox`와 `추가 이용거리(M)_BoxCox` 컬럼을 활용해 정규성 검정과 통계 분석을 진행할 예정입니다.

### 5.7.2. 정규성 검정

```python
# Q-Q 플롯: 추가 이용시간과 추가 이용거리의 정규성 시각화
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 추가 이용시간(분)_BoxCox Q-Q 플롯
stats.probplot(box_cox_df['추가 이용시간(분)_BoxCox'], dist="norm", plot=axes[0])
axes[0].set_title("Q-Q 플롯: 추가 이용시간(분)_BoxCox")

# 추가 이용거리(M)_BoxCox Q-Q 플롯
stats.probplot(box_cox_df['추가 이용거리(M)_BoxCox'], dist="norm", plot=axes[1])
axes[1].set_title("Q-Q 플롯: 추가 이용거리(M)_BoxCox")

plt.tight_layout()
plt.show()

# Kolmogorov-Smirnov Test (K-S Test): 정규성 검정
print("📊 Kolmogorov-Smirnov Test 결과")
for col in ['추가 이용시간(분)_BoxCox', '추가 이용거리(M)_BoxCox']:
    stat, p = stats.kstest(box_cox_df[col], 'norm', args=(box_cox_df[col].mean(), box_cox_df[col].std()))
    print(f"{col} - K-S 통계량: {stat:.4f}, p-value: {p:.4e}")

    if p > 0.05:
        print(f"✅ {col}: 정규성을 만족합니다. (p > 0.05)\n")
    else:
        print(f"❌ {col}: 정규성을 만족하지 않습니다. (p <= 0.05)\n")
```

1. Q-Q 플롯 분석:
    - `추가 이용시간(분)_BoxCox`:
        - 그래프를 보면 중앙 부분은 직선에 근접하지만, 양쪽 끝부분(꼬리 부분)에서 심하게 벗어나는 모습을 확인할 수 있습니다.
        - 이는 정규성을 완전히 확보하지 못했음을 시사합니다.
    - `추가 이용거리(M)_BoxCox`:
        - 이 역시 중앙부에서는 직선과 유사하지만, 양쪽 꼬리 부분에서 곡선을 그리며 이탈하고 있습니다.
        - 정규성이 불완전한 상태임을 나타냅니다.
    - 결론:
        - Box-Cox 변환으로 일부 정규성이 개선되었으나, 완전히 정규성을 확보하지는 못한 상태입니다.

2. Kolmogorov-Smirnov (K-S) Test 분석
    - `추가 이용시간(분)_BoxCox`:
        - K-S 통계량: 0.0391
        - p-value: 0.0000
        - 결론: p-value가 0.05 미만으로, 정규성을 기각합니다.
            - → 즉, Box-Cox 변환 후에도 정규성을 만족하지 않음을 의미합니다.
    - `추가 이용거리(M)_BoxCox`:
        - K-S 통계량: 0.0099
        - p-value: 0.0000
        - 결론: p-value가 0.05 미만, 정규성을 기각합니다.
            - → 마찬가지로 정규성을 만족하지 않음을 의미합니다.

3. 최종 해석 및 다음 단계
    - Box-Cox 변환을 통해 데이터의 왜도를 일정 부분 완화했으나, 정규성 확보에는 실패했습니다.
    - 정규성 기반 분석(ANOVA, Tukey's HSD 등)을 진행할 수 있지만, 결과 해석에 한계가 있을 수 있습니다.
    - 따라서, 비정규성에도 강건한 방법인 Kruskal-Wallis Test와 Dunn's Test의 신뢰도가 더 높다고 판단됩니다.

4. 다음 단계
    - 정규성 확보에 실패했지만, 데이터의 이상치가 정규성에 부정적 영향을 미쳤을 가능성을 고려해 Winsorization을 진행할 예정입니다.
    - Winsorization은 극단값을 일정한 임계값으로 조정하여 데이터의 분포를 안정화하는 방법입니다.
    - 이를 통해 정규성 개선 여부를 재확인하고, 변환된 데이터를 기반으로 ANOVA와 Tukey's HSD 사후검정을 수행할 예정입니다.
    - 만약 Winsorization 이후에도 정규성이 충분히 확보되지 않는다면, 비모수적 분석(Welch ANOVA, Kruskal-Wallis Test)의 결과를 최종 결론으로 채택할 것입니다.

### 5.7.3. Winsorization (극단값 조정)

```python
# Winsorization 비율 설정 (상하위 5% 조정)
lower_percent = 0.05
upper_percent = 0.05

# 추가 이용시간(분)과 추가 이용거리(M) 변수에 Winsorization 적용
box_cox_df['추가 이용시간(분)_Winsorized'] = winsorize(box_cox_df['추가 이용시간(분)_BoxCox'], limits=(lower_percent, upper_percent))
box_cox_df['추가 이용거리(M)_Winsorized'] = winsorize(box_cox_df['추가 이용거리(M)_BoxCox'], limits=(lower_percent, upper_percent))

# Winsorization 결과 확인
box_cox_df[['연령대', '추가 이용시간(분)_BoxCox', '추가 이용시간(분)_Winsorized', '추가 이용거리(M)_BoxCox', '추가 이용거리(M)_Winsorized']].head(3)
```

- Box-Cox 변환을 통해 데이터의 왜도를 일정 부분 완화했지만, 정규성 확보에는 실패했습니다. 이는 데이터 내 여전히 존재하는 극단값(outlier)들이 분석에 영향을 미치고 있다는 것을 시사합니다.
- 따라서, Winsorization 기법을 적용하여 상·하위 5%의 극단값을 조정했습니다. Winsorization은 극단값을 완전히 제거하는 대신, 극단적인 값을 일정 임계값으로 조정하여 분석의 왜곡을 줄이면서도 전체 데이터의 구조를 유지하는 방법입니다.
- 이제 Winsorization이 적용된 데이터를 기반으로 다시 정규성 검정(Q-Q 플롯, K-S Test)을 수행하여 데이터 분포가 분석에 적합하게 조정되었는지 확인합니다.

### 5.7.4. 정규성 검정 (Winsorized)

```python
# 정규성 검정 대상 컬럼
time_col = '추가 이용시간(분)_Winsorized'
distance_col = '추가 이용거리(M)_Winsorized'

# Q-Q 플롯 그리기
plt.figure(figsize=(14, 6))

# 추가 이용시간(분) Q-Q 플롯
plt.subplot(1, 2, 1)
stats.probplot(box_cox_df[time_col], dist="norm", plot=plt)
plt.title(f'Q-Q 플롯: {time_col}')

# 추가 이용거리(M) Q-Q 플롯
plt.subplot(1, 2, 2)
stats.probplot(box_cox_df[distance_col], dist="norm", plot=plt)
plt.title(f'Q-Q 플롯: {distance_col}')

plt.tight_layout()
plt.show()

# Kolmogorov-Smirnov (K-S) Test
time_ks_stat, time_ks_p = stats.kstest(box_cox_df[time_col], 'norm', args=(box_cox_df[time_col].mean(), box_cox_df[time_col].std()))
distance_ks_stat, distance_ks_p = stats.kstest(box_cox_df[distance_col], 'norm', args=(box_cox_df[distance_col].mean(), box_cox_df[distance_col].std()))

# 결과 출력
print("\n📊 Kolmogorov-Smirnov Test 결과")
print(f"{time_col} - K-S 통계량: {time_ks_stat:.4f}, p-value: {time_ks_p:.4f}")
print(f"{distance_col} - K-S 통계량: {distance_ks_stat:.4f}, p-value: {distance_ks_p:.4f}")

# 정규성 여부 판정
alpha = 0.05
print("\n✅ 정규성 여부 판정:")
print(f"{time_col}: {'✅ 정규성 만족' if time_ks_p > alpha else '❌ 정규성 불만족'}")
print(f"{distance_col}: {'✅ 정규성 만족' if distance_ks_p > alpha else '❌ 정규성 불만족'}")
```

1. Q-Q 플롯 분석:
    - `추가 이용시간(분)_Winsorized`: Q-Q 플롯에서 데이터가 대체로 직선에 근접하나, 분포의 끝 부분(꼬리)에서 직선을 벗어나는 모습이 보입니다. 이는 여전히 정규성에서 벗어난 상태임을 시사합니다.
    - `추가 이용거리(M)_Winsorized`: 마찬가지로 데이터의 중심부는 직선과 유사하게 정렬되었지만, 양쪽 끝부분에서 정규성에서 이탈하는 패턴이 나타납니다.

2. Kolmogorov-Smirnov (K-S) Test
    - `추가 이용시간(분)_Winsorized`: 통계량 = 0.0457, p-value = 0.0000
    - `추가 이용거리(M)_Winsorized`: 통계량 = 0.0386, p-value = 0.0000
        - → 두 변수 모두 p-value가 0.05 미만이므로, 귀무가설(정규성을 따른다)을 기각합니다.

3. 결론 및 해석
    - Winsorization을 통해 이상치를 완화했음에도 불구하고, 데이터는 여전히 정규성을 만족하지 않는 것으로 나타났습니다.
    - 이는 데이터가 여전히 비정규적 특성을 유지하고 있으며, 정규성을 전제로 하는 ANOVA와 Tukey HSD 분석을 수행할 경우, 결과 해석에 주의가 필요함을 의미합니다.
    - 따라서, 비모수적 접근법인 Kruskal-Wallis Test와 Dunn's Test가 여전히 신뢰성이 높은 분석 방법으로 간주됩니다.

    - 실질적 차이: 절대적인 차이는 크지 않지만, 노년층의 실패율이 다소 높아 이용 과정에서의 장벽 가능성을 시사합니다.
    - 추가 분석 권장: 단순 비율 차이 외에도 로지스틱 회귀를 통해 연령과 실패율 간의 관계를 분석하면, 다른 요인의 영향을 통제한 상태에서 연령의 독립적 영향을 확인할 수 있습니다.
- 다음 단계:
    - 다변량 분석을 통해 연령과 실패율 간의 관계를 보다 명확히 검증할 필요가 있습니다. 이를 위해, 다음 단계로 로지스틱 회귀 분석(Logistic Regression)을 진행하여, 연령이 다른 변수들을 통제한 상태에서도 실패율을 유의하게 설명하는지를 분석할 예정입니다.

## 6.3. 로지스틱 회귀 분석(Logistic Regression)

```python
# 이용 실패 여부 컬럼 추가
access_df['이용 실패 여부'] = ((access_df['대여 대여소번호'] == access_df['반납 대여소번호']) & (access_df['이용거리(M)'] == 0)).astype(int)

# 연령대 변수 라벨 인코딩 (청년: 0, 중년: 1, 노년: 2)
le = LabelEncoder()
access_df['연령대_코드'] = le.fit_transform(access_df['연령대'])

# 독립 변수(X)와 종속 변수(y) 설정
X = access_df[['연령대_코드']]
X = sm.add_constant(X)  # 상수항 추가
y = access_df['이용 실패 여부']

# 로지스틱 회귀 모델 생성
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# 결과 요약 출력
print(result.summary())

# 오즈비(odds ratio) 계산
odds_ratios = pd.DataFrame({
    '변수': X.columns,
    '오즈비': result.params.apply(lambda x: round(np.exp(x), 2)),
    'p-value': result.pvalues
})

print("\n=== 오즈비 (Odds Ratio) ===")
print(odds_ratios)
```

### 6.3.1. 주요 분석 과정
1. 이용 실패 여부: `대여 대여소번호`와 `반납 대여소번호`가 같고, `이용거리(M)`가 `0`인 경우를 실패로 간주해 `1`로 처리합니다.
2. 연령대 인코딩: 청년(`0`), 중년(`1`), 노년(`2`)로 라벨을 부여해 회귀 분석이 가능하도록 변환합니다.
3. 로지스틱 회귀 분석: 연령대가 이용 실패 여부에 미치는 영향을 분석합니다.
4. 오즈비 해석: 오즈비가 1보다 크면 실패 확률이 높아지고, 1보다 작으면 낮아짐을 의미합니다.
    - 노년층의 오즈비가 높을 경우, 청년층 대비 실패 확률이 증가한 것임을 시사합니다.

    - 이용 실패율은 대여 시도 후 실패한 경우만을 분석 대상으로 삼아, 대여 시도 자체를 하지 않은 집단은 고려하지 않았습니다.
    - 이는 정보 접근성 부족으로 인해 대여 시도조차 하지 않은 노년층의 현실을 충분히 반영하지 못했을 가능성이 있습니다.
3. 환경적 요인 미반영:
    - 분석 과정에서 따릉이 이용과 관련된 기상, 도로 환경, 안전 인프라와 같은 외부 요인은 반영되지 않았습니다.
    - 노년층의 자전거 이용은 날씨, 교통안전 인식, 자전거 도로 접근성과도 밀접하게 관련되어 있으므로, 외부 환경 요인을 포함한 분석이 보완되어야 합니다.

## 7.4. 결론 및 향후 연구 방향
본 연구를 통해, 노년층의 낮은 따릉이 이용률은 인지 부족과 정보 접근성의 한계에서 비롯된 것임을 확인했습니다. 이는 노년층이 따릉이에 대해 충분히 알고, 쉽게 접근할 수 있다면 이용률이 자연스럽게 증가할 가능성이 크다는 의미입니다.

향후 연구에서는 다음과 같은 방향으로 분석을 확장하는 것이 필요합니다.

1. 비이용자 대상 설문조사:
    - 따릉이를 이용하지 않은 노년층을 대상으로 인지 수준, 심리적 장벽, 이용 의사 등을 조사하고, 데이터 기반의 정책적 접근을 보완합니다.
2. 환경적 요인 분석:
    - 기상, 자전거 도로 접근성, 교통안전 인식 등 외부 요인이 따릉이 이용률에 미치는 영향을 분석하고, 보다 정교한 정책 설계를 지원합니다.
3. 정책 시행 후 효과 검증:
    - 제안된 정책이 시행된 후, 노년층의 따릉이 이용률과 이용 과정에서의 불편 사항을 지속적으로 모니터링하고, 정책의 실질적 효과를 평가하는 과정이 필요합니다.

결론적으로, 따릉이 이용률 증진을 위해서는 인프라 확충과 함께 정보 접근성을 높이는 정책적 접근이 필수적입니다. 서울시와 지역 사회가 협력하여 노년층이 따릉이를 안전하고 편리하게 이용할 수 있도록 환경을 조성하는 것이 중요합니다.
